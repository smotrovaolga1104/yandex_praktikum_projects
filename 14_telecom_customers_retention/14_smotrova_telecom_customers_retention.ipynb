## Введение

### Цель проекта

Необходимо построить прототип модели, позволяющей предсказывать уход клиента на основании его персональных данных и условий его договора с провайдером. Модель должна достигать значения *ROC AUC* не менее 0.85. Проект должен быть завершен в течение двух недель с момента постановки задачи Заказчиком.

Использование модели позволит уменьшить маркетинговые расходы за счет более таргетированной работы по удержанию потенциально уходящих клиентов и экономии на привлечении новых пользователей.

Для реализации поставленной задачи следует:

1. Проанализировать предоставленные данные на предмет их полноты, релевантности, при необходимости устранить проблемы в данных;
2. Выбрать архитектуры моделей, наиболее подходящие для имеющихся данных;  
3. Подготовить данные для дальнейшего обучения модели: удалить нерелевантные данные, выделить из общего массива обучающую и тестовую выборки, при необходимости провести масштабирование, кодировку категориальных признаков и т.п.;
4. Обучить несколько типов моделей с подбором гиперпараметров для максимизации метрики качества;
5. На основании результатов кросс-валидации выбрать наилучшую модель, проверить ее качество на тестовой выборке;
6. Если выбранная модель не достигнет требуемого уровня качества, повторить п. выше, пока требуемый порог качества не будет достигнут.

### Объединение и унификация данных

````python
# Объединяем все файлы в один по индентификатору клиента:
df = contracts.merge(pers_info, on='customerID', how='outer')
df = df.merge(internet, on='customerID', how='outer', copy=False)
df = df.merge(phone, on='customerID', how='outer', copy=False)
df.info()
````

Заметим, что при использованном типе объединения число строк в полученном датафрейме совпало с общим числом строк в файлах о контрактах и персональной информации. Число ненулевых значений в параметрах интернет-услуг и телефонии совпадает с исходным числом строк в соответствующих файлах. Т.е. в этих файлах одна и та же выборка по клиентам.

Теперь, кода данные объединены, образовались пропуски в данных об услугах интернета и телефонии. с наибольшей вероятностью отсутсвие информации о подключенных услугах говорит о том, что клиент ими не пользуется. Поэтому для признаков, отвечающих за наличие услуги как таковой (`InternetService`, `MultipleLines`) заменим пропущенные значения на **Disabled**, чтобы выделить тех клиентов, которые не подключали услуги.

Оставшиеся пропуски заменим на "No" для интернет-услуг, т.к. отсутсвие интернет-подключения автоматически означает отсутствие доп.услуг. Т.е. введение дополнительного значения (напр., no_data) избыточно.

````python
# Заполняем пропуски для основных услуг:
df['InternetService'] = df['InternetService'].fillna('Disabled')
df['MultipleLines'] = df['MultipleLines'].fillna('Disabled')
# Заполняем оставшиеся пропуски:
df = df.fillna('No')
````

````python
# Для дальнейшей визуализации:
def graph(var, ax=None, xlabel='', title=''):
    pic = sns.histplot(x=var,
                      hue='exited',
                      data = df,
                      ax = ax,
                      multiple='dodge',
                      common_norm = False,
                      stat='probability')
    if ax==None:
        plt.ylabel('Доля в группе')
        plt.xlabel(xlabel)
        plt.title(title)
    else:
        ax.set_ylabel('Доля в группе')
        ax.set_xlabel(xlabel)
        ax.set_title(title)
    return pic
````

Заметим, что даты округлены до месяца (скорее всего, данные агрегируются по месяцам на уровне базы данных). При этом уникальных дат окончания (если есть) всего 4 против 77 дат начала и захватывают лишь 4 последних месяца (выгрузка на 1 ферваля 2020 г.), хотя данные о заключении договора начинаются с октября 2013 г.
 
Данные о начале и окончании контракта в виде дат сами по себе неприменимы для модели. Однако, по дате начала можно судить о длительности отношений клиента с провайдером, а наличие даты окончания говорит о том, что клиент расторг договор. Введем два синтетических параметра: срок действия договора и факт расторжения контракта. Поскольку известно, что данные актуальны на 1 февраля 2020 г., будем отсчитывать срок действия договора (в днях) от этой даты.

````python
# Создадим признак разрыва договора:
df['exited'] = df['end_date'].where(df['end_date']=='No', 1)
df['exited'] = df['exited'].replace('No', 0)
df.loc[df['exited']==0, 'end_date'] = None  # для дальнейшей конвертации

# Создадим признак длительности договора:
df['begin_date'] = pd.to_datetime(df['begin_date']) # конвертируем формат даты
df['end_date'] = pd.to_datetime(df['end_date'])
threshold_date = pd.to_datetime('2020-02-01')

df.loc[df['exited']==0, 'duration'] = pd.to_numeric(
    (threshold_date - df['begin_date']).dt.days) # находим длительность действующих контрактов
df.loc[df['exited']==1, 'duration'] = pd.to_numeric(
    (df['end_date'] - df['begin_date']).dt.days) # находим длительность расторгнутых
````

````python
exited_share = sum(df['exited'])/len(df)
print(f'Доля ушедших клиентов в выборке составляет {exited_share:.2%}.')

# Распределения длительности:
graph('duration', xlabel='В днях', title='Длительность договора')
plt.show()
````

**Наблюдения:**
- Выборка несбалансирована по классам, что нужно будет учесть при выделении тестовой выборки.
- Среди ушедших клиентов существенно выше доля тех, чей договор длился до 100 дней. И наоборот, среди оставшихся существенно выше доля тех, кто пользуется услугами провайдера более 5 лет. Возможно, это связано с неравномерностями в дате расторжения договора. На "серединном" участке распределений существенных различий не заметно.

### Парная корреляция

Проанализируем статистические зависимости между параметрами. Поскольку параметры разнородные по типу (бинарные, категориальные, количественные), стандартный коэффициент корреляции Пирсона не применим. Для анализа статистических зависимостей будем использовать коэффициент $\phi_K$, позволяющий работать с разными типами переменных, а также чуствительный к нелинейным зависимостям.

````python
int_cols = ['monthly_charges', 'duration']
# Расчет коэффициентов:
corr_table = df.phik_matrix(interval_cols = int_cols)

# Визуализация:
report.plot_correlation_matrix(corr_table.values, 
                              x_labels=corr_table.columns, 
                              y_labels=corr_table.index, 
                              vmin=0, vmax=1,
                              color_map = 'Blues',
                              title="Коэфициенты парной корреляции", 
                              figsize=(10, 8))
plt.tight_layout()
````

````python
# Матрица значимости:
corr_significance = df.significance_matrix(interval_cols = int_cols)
report.plot_correlation_matrix(corr_significance.values, 
                              x_labels=corr_significance.columns, 
                              y_labels=corr_significance.index,
                              vmin = 5, vmax = -5,
                              color_map = 'Blues',
                              title="Значимость коэффициентов корреляции", 
                              figsize=(10, 8))
plt.tight_layout()
````

**Наблюдения по целевому признаку:**
- Наибольшую статистическую взаимосвязь с фактом ухода демонстрируют срок действия договора и способ оплаты (коэффициенты 0.47 и 0.45); взаимосвязь, тем не менее, средней силы.
- Тип договора, размер ежемесячных расходов и получение электроного чека демонстрируют средне-слабую (0.25-0.36) статистически значимую связь с целевым признаком;
- Услуги тех.поддержки и нежелательной блокировки сайтов оказывают средне-слабое (ок. 0.25), однако статистически значимое влияние на разрыв договора;
- Услуги по стримингу ТВ и фильмов, а также антивирус и облачное хранение оказывают более слабое (ок. 0.1), однако также статистически значимое (на уровне 5 стандартных отклонений) влияние на факт ухода клиента.
- Наличие супруга/(и), иждивенцев и пенсионный вознаст имеют статистически значимое средне-слабое влияние на целевой признак, в то время как пол клиента статистически не влияет на факт ухода (подтвердили вывод исследовательского анализа).

**Наблюдения по объясняющим признакам:**
- Ежемесячные расходы сильнее всего скоррелированы со всеми остальными признаками, кроме пола: сильнейшая взаимосвязь (0.92) наблюдается с типом интернет-подключения, также сильная взаимосвязь (0.83) с услугами стриминга ТВ и фильмов. Это логично, т.к. добавление любой услуги увеличивает величину расходов.
- Тип услуг телефонии слабо связан с большинством признаков (включая целевой), кроме интернет-услуг и ежемесячных расходов. Рассмотрим взаимосвязь с типом интернет-услуг подробнее:

````python
var_1 = 'multiple_lines'
var_2 = 'internet_service'

outlier_signifs, binning_dict = df[[var_1,var_2]].outlier_significance_matrix(retbins=True)

zvalues = outlier_signifs.values
xlabels = binning_dict[var_2] if var_2 in binning_dict.keys() else outlier_signifs.columns
ylabels = binning_dict[var_1] if var_1 in binning_dict.keys() else outlier_signifs.index

report.plot_correlation_matrix(zvalues, x_labels=xlabels, y_labels=ylabels, 
                              x_label=var_2, y_label=var_1,
                              vmin=-5, vmax=5, 
                              identity_layout=False, 
                              figsize=(5, 5))
````

Чаще всего эти два признака встречаются в определенных комбинациях. Для уменьшения мультиколлинеарности сгегерируем признак, обозначающий комбинацию основных услуг провайдера, вместо двух раздельных признаков.
- Взаимосвязи между дополнительными услугами статистически значимы и средне-сильные. Кроме того, них похожий профиль взаимосвязей с остальными полями, кроме целевого признака (см.выше). Для уменьшения мультиколлинеарности объединим все услуги, кроме тех.поддержки и блокировки сайтов в прочие услуги (`other_services`). Поле будет принимать значение *Yes*, если присутствует хотя бы одна из исходных услуг.
- Наблюдается средне-сильная корреляция (0.65) между наличием супруга/(и) и иждивенцев, что логично (наличие детей с большоей вероятностью означает наличие партнера).
- Тип договора средне-сильно связан с длительностью договора, что также объяснимо: краткосрочный договор легче разорвать.

## Исследование моделей

Рассмотрим три типа ансамблевых моделей: случайный лес, а также два варианта градиентного бустинга - *CatBoost* и *LightGBM*. Все они основаны на решающих деревьях, поэтому мультиколлинеарность не ухудшит качество моделей. Кроме того, для единственного оставшегося числового признака (длительность договора) не требуется масштабирование. 

Поскольку датасет не слишком большой (ок. 7 тыс. объектов), время обучения случайного леса не будет слишком большим по сравнению с моделями градиентного бустинга.

Модели требуют разной подготовки данных:
- *CatBoost* способен работать с категориальными признаками напрямую,
- Для случайного леса и *LightGBM* подойдет порядковое кодирование, т.к. *LightGBM* работает с категориальными признаками в формате целых числел. Кроме того, порядковоое кодирование предпочтительнее прямого кодирования (One-Hot Encoding), поскольку уменьшает возможную глубину дерева.

### Подготовка выборок

Прежде, чем кодировать данные, выделим из датасета обучающую и тестовую выборки. Стратифицируем выборки по целевому признаку.

````python
features = df_clean.drop(['exited'], axis=1)
target = df['exited']

# Выделяем тестовую выборку:
features_train, features_test, target_train, target_test = train_test_split(features,
                                                                           target,
                                                                           test_size = 0.25,
                                                                           stratify = target,
                                                                           random_state = RANDOM_STATE)
print('Размер обучающей выборки:', features_train.shape)
print('Размер тестовой выборки:', features_test.shape)
````

Чтобы не готовить два отдельных датасета для разных моделей, воспользуемся инструментом `pipeline`. Кроме того, использование этого инструмента позволит избежать "утечки" кодирования в валидационную выборку (формируется из обучающей средствами `GridSearchCV`).

````python
# Названия категориальных колонок для CatBoost:
cat_features = list(features.columns)
cat_features.remove('duration')

# Индексы категориальных колонок:
cat_cols = []
for i in range(features.shape[1]):
    if features.columns[i] in cat_features:
        cat_cols.append(i)
        
# Инициализируем кодировщик для катег.переменных, чтобы использовать его в pipeline:
encoder = ColumnTransformer(transformers=[('encoder', OrdinalEncoder(), cat_cols)],
                            remainder='passthrough')
````

### Обучение моделей

Обучим разные типы моделей с подбором гиперпараметров. Будем использовать встроенное балансирование классов в моделях.

````python
# Задаем используемые модели:
pipe_lgbm = make_pipeline(encoder, 
                         LGBMClassifier(class_weight='balanced',
                                       random_state = RANDOM_STATE))
pipe_rf = make_pipeline(encoder, 
                       RandomForestClassifier(class_weight='balanced',
                                             random_state = RANDOM_STATE))
cb = CatBoostClassifier(loss_function='Logloss',
                       cat_features = cat_features,
                       class_weights=[exited_share, 1-exited_share],
                       verbose=False,
                       random_seed = RANDOM_STATE)

ests = {'Бустинг CatBoost' : cb,
        'Бустинг LGBM' : pipe_lgbm,
        'Случайный лес' : pipe_rf}
````

````python
# Задаем параметры для подбора:
params = {'Бустинг CatBoost' : {'iterations' : [100, 200, 300],
                               'learning_rate' : [0.3, 0.5, 0.7],
                               'depth' : range(2, 11)},
          'Бустинг LGBM' : {'lgbmclassifier__n_estimators' : [100, 200, 300],
                            'lgbmclassifier__learning_rate' : [0.3, 0.5, 0.7],
                            'lgbmclassifier__max_depth' : range(2, 11)},
          'Случайный лес' : {'randomforestclassifier__n_estimators' : [50, 100, 200],
                             'randomforestclassifier__max_depth' : range(2, 11),
                             'randomforestclassifier__max_features' : ['sqrt', 'log2', None]}
         }
# Заготовка под результаты валидации:
results = {'Бустинг CatBoost' : None,
           'Бустинг LGBM' : None,
           'Случайный лес' : None}
# Заготовка под лучшие модели:
best_models = {'Бустинг CatBoost' : None,
               'Бустинг LGBM' : None,
               'Случайный лес' : None}
# Заготовка под лучшие значения метрики:
best_scores = pd.DataFrame(index = ests.keys())
````

````python
%%time
# Обучаем модели с подбором гиперпараметров:
for i in ests.keys():
    grid = GridSearchCV(estimator = ests[i],
                       param_grid = params[i],
                       scoring = 'roc_auc',
                       refit=True)
    grid.fit(features_train, target_train)
    results[i] = grid.cv_results_
    best_models[i] = grid.best_estimator_
    best_scores.loc[i, 'ROC AUC'] = grid.best_score_
````

### Выбор модели

Рассмотрим результаты обучения моделей:

````python
for i in ests.keys():
    best_scores.loc[i, 'F1'] = f1_score(target_train, best_models[i].predict(features_train))
    best_scores.loc[i, 'Accuracy'] = best_models[i].score(features_train, target_train)

print('Результаты лучших моделей своего типа:')
display(best_scores)

print('Полная таблица результатов кросс-валидации:')
visual_table = pd.DataFrame()
visual_table['Параметры для бустинга'] = results['Бустинг CatBoost']['params']
visual_table['CatBoost'] = results['Бустинг CatBoost']['mean_test_score']
visual_table['LGBM'] = results['Бустинг LGBM']['mean_test_score']
visual_table['Параметры для случ.леса'] = results['Случайный лес']['params']
visual_table['Случайный лес'] = results['Случайный лес']['mean_test_score']
display(visual_table)
````

Все три *лучшие модели своего типа* преодолели требуемый порог качества по метрике *ROC AUC* (0.85). Наивысшей метрики достигла модель градиентного бустинга, *CatBoost*, показатель *LightGBM* немного хуже (ок. 0.940 и 0.927 соответственно). Вспомогательные метрики (F1-мера и Accuracy) у моделей бустинга также похожи. Модель случайного леса показала наихудший из трех результатов по основной метрике (ок. 0.87).

Наибольший вклад в объяснение целевого признака вносит *длительность договора* - относительная важность более 80%. *Маркер основных услуг* и *тип договора* внесли ок. 6% и 8% соответственно. Индивидуальный вклад остальных признаков не превышает 2%.
Вероятно, смещение важности признаков в пользу длительности договора связано с особенностями формирования исходного датасета (см. выше); велика вероятность, что качество модели и вклады других объясняющих признаков увеличатся при использовании однородных исходных данных.

````python
importances['category'] = importances['feature_name'].where(importances['value'] > 4, 'other')
pivot = importances.pivot_table(index='category', values='value', aggfunc='sum')
plt.pie(x=pivot['value'], 
       labels = pivot.index,
       autopct="%.1f%%");
````

Предлагаемая модель пропускает ок. 15,5% действительно уходящих клиентов ("пропущенные"); при этом примерно 25,5% клиентов, классифицированных как уходящие, в действительности не разорвут договор ("избыточные"). Такое разделение основано на пороге классификации в 50% (если предсказанная вероятность ухода больше порога - клиент идентифицируется моделью как уходящий). Точность решений на основе модели можно повысить за счет изменения порога; его оптимальный уровень будет определяться балансом между стоимостью ухода "пропущенных" клиентов и дополнительными маркетинговыми расходами на "излишних" клиентов.

Трудности, возникшие в процессе работы:
